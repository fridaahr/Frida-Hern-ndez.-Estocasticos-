## Impotar librerías 

Importamos las librerías con:

```python
import sympy as sp
import networkx as nx
import matplotlib.pyplot as plt
```

## Crear matrices en Sympy

En Sympy podemos definir matrices utilizando:

```python
A = sp.Matrix([[0,0,2],
               [3,6,0],
               [0,1,3]])
```

Imprimimos la matriz:

```python
A
```

## Crear un vector

En sympy podemos crear un vector utilizando:

```python
v = sp.Matrix([4,0,6])
```

Imprimimos el vector:

```python
v
```

## Multiplicación matriz - vector

Un vector puede multiplicarse por una matriz siempre que las dimensiones sean compatibles.

Para multiplicar matriz - vector, utilizamos:

```python
resultado = A*v
```

Imprimimos el resultado:

```python
resultado
```

## Potencias de matrices

También es posible calcular potencias de matrices utilizando:

```python
A2 = A**2
```

El operador `**` se utiliza en Python para calcular potencias.

Imprimimos el resultado de la potencia:

```python
A2
```

## Solución de sistemas de ecuaciones

Sympy también permite resolver sistemas de ecuaciones lineales utilizando estos pasos:

```python
x,y = sp.symbols('x y')
```

La función `symbols()` se utiliza para crear variables simbólicas. En este caso estamos creando las variables `x` y `y`

```python
eq1 = sp.Eq(2*x + y,5)
```

Aquí definimos la primera ecuación del sistema: 2x + y = 5.

```python
eq2 = sp.Eq(x - y,1)
```

Aquí definimos la segunda ecuación del sistema: x - y = 1.

```python
solucion = sp.solve((eq1,eq2),(x,y))
```

La función `solve()` se utiliza para encontrar la solución del sistema de ecuaciones. Indicando las ecuaciones que queremos resolver y las variables que buscamos.

Imprimimos el resultado:

```python
solucion
```

## Valores propios y vectores propios

También podemos calcular valores y vectores propios utilizando:

```python
A.eigenvals()
```

La función `eigenvals()` calcula los valores propios de la matriz.

El resultado se muestra como un diccionario donde:

• La clave representa el valor propio.

• El valor indica su multiplicidad (cuántas veces aparece).

```python
A.eigenvects()
```

La función `eigenvects()` calcula los vectores propios de la matriz.

El resultado contiene tres elementos:

• El valor propio.

• Su multiplicidad.

• El vector propio correspondiente.

## Diagramas de transición con NetworkX

Podemos mostrar diagramas de transición usando NetworkX con:

```python
G = nx.DiGraph()
```

Creamos un grafo dirigido usando la función `DiGraph()`

Agregamos una conexión desde el nodo A hacia el nodo B:

```python
G.add_edge("A","B")
```

Agregamos una conexión desde el nodo B hacia el nodo C:

```python
G.add_edge("B","C")
```

Agregamos una conexión desde el nodo C hacia el nodo A:

```python
G.add_edge("C","A")
```

Ahora, la función `draw()` dibuja el grafo.

El parámetro `with_labels=True` indica que queremos mostrar los nombres de los nodos:

```python
nx.draw(G, with_labels=True)
```

En algunos casos el grafo se imprime solo con la línea anterior, en caso contrario, el grafo se imprime utilizando `plt.show()`

## Crear un diagrama de transición a partir de una matriz

También es posible construir un grafo directamente a partir de una matriz.

Convertiremos la matriz (A) a formato Numpy:

```python
import numpy as np

matriz_np = np.array(A).astype(int)
```

NetworkX trabaja de forma más directa con arreglos de NumPy.

Por esta razón convertimos la matriz de Sympy a un arreglo de NumPy utilizando `np.array()`.

El método `astype(int)` asegura que los valores se manejen como números enteros.

Ahora, creamos el grafos apartir de la matriz utilizando:

```python
G = nx.from_numpy_array(matriz_np, create_using=nx.DiGraph)
```

La función `from_numpy_array()` permite crear un grafo directamente a partir de una matriz.

Cada fila y columna de la matriz representa un nodo del grafo.

El parámetro `create_using=nx.DiGraph` indica que el grafo será dirigido, es decir, que las conexiones tienen dirección.

Finalmente, dibujamos el grafo:

```python
nx.draw(G, with_labels=True)
plt.show()
```

La función `draw()` dibuja el grafo generado.

El parámetro `with_labels=True` permite mostrar las etiquetas de los nodos.

La función `plt.show()` muestra el gráfico en pantalla.
